**Standard Scaling vs Minâ€“Max Scaling**

Standard Scaling transforms features to have a mean of 0 and a standard deviation of 1. It is also less sensitive to small variations but can be influenced by outliers.

Minâ€“Max Scaling rescales features to a fixed range, usually [0,1]. It is preferred when the data does not follow a normal distribution. However, it is highly sensitive to outliers, as extreme values affect the entire scale.

## **One-Hot Encoding** represents each category as a separate binary feature, preventing false ordinal relationships.
Example: Color = {Red, Blue, Green} - Color_Red, Color_Blue, Color_Green.
It is ideal for nominal data like gender or country but increases the number of features.

# **Label Encoding** assigns an integer to each category.
Example: Size = {Small, Medium, Large} - {Small=1, Medium=2, Large=3}.
It is suitable for ordinal data where order matters, but can mislead models if used on nominal categories.

# Feature Engineering

```python
new_df.head()
```

```python
new_df['net_capital'] = new_df['capital_loss'] - new_df['capital_gain']
```

Rationale:

* Capital gain and capital loss are closely related financial variables.

* Combining them into a single net capital outcome better represents overall investment income.

* This reduces redundancy and may improve the modelâ€™s ability to capture wealth-related effects on income.

```python
def is_overtime(x):
  if x >= 40:
    return 1
  else:
    return 0

new_df['overtime'] = new_df['hours_per_week'].apply(is_overtime)
#data.drop('hours_per_week',axis=1,inplace=True)
new_df.head()
```

Rationale:

Individuals working more than standard full-time hours often have higher earnings. This binary feature simplifies the relationship between working hours and income and helps the model distinguish between regular and overtime workers.

```python
import numpy as np


new_df['capital_gain_log'] = np.log1p(pd.to_numeric(new_df['capital_gain'], errors='coerce').fillna(0))
new_df['has_capital_gain'] = (new_df['capital_gain'] > 0).astype(int)
new_df
```

Justification

* Compresses extreme values in the long tail

* Preserves zero values (since log(1 + 0) = 0)

* Produces a more balanced distribution with reduced skewness

* Helps models better capture relative differences among non-zero values

* has_capital_gain Column is useful to identifiy the capital_gains(binary indicator)

The Isolation Forest algorithm was used to identify and remove outliers by isolating observations that are few and different from the majority of the data. It is efficient for high-dimensional datasets and does not assume any specific data distribution.

Outliers can negatively affect model performance by skewing feature distributions, biasing parameter estimates, and reducing prediction accuracy, especially for distance-based and linear models. Removing outliers helps improve model stability, generalization, and overall predictive performance.
